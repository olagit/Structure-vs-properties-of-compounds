```python
import numpy as np

maccs2 = maccs.copy()
maccs2['Donor'] = np.zeros(maccs.shape[0], dtype=int)
maccs2['Acceptor'] = np.zeros(maccs.shape[0], dtype=int)
maccs2['Aromatic'] = np.zeros(maccs.shape[0], dtype=int)
maccs2['Hydrophobe'] = np.zeros(maccs.shape[0], dtype=int)
maccs2['LumpedHydrophobe'] = np.zeros(maccs.shape[0], dtype=int)

maccs2.head()
```

```python
from rdkit import Chem
from rdkit.Chem import ChemicalFeatures
from rdkit import RDConfig
```

```python
import os
fdefName = os.path.join(RDConfig.RDDataDir,'BaseFeatures.fdef')
factory = ChemicalFeatures.BuildFeatureFactory(fdefName)
```

```python
for i in range(maccs.shape[0]):
    m = Chem.MolFromSmiles(maccs['smiles'][i])
    feats = factory.GetFeaturesForMol(m)
    if feats:
        for j in range(len(feats)):
            if feats[j].GetFamily() == 'Donor':
                maccs2.loc[i, 'Donor'] += 1
            elif feats[j].GetFamily() == 'Acceptor':
                maccs2.loc[i, 'Acceptor'] += 1
            elif feats[j].GetFamily() == 'Aromatic':
                maccs2.loc[i, 'Aromatic'] += 1
            elif feats[j].GetFamily() == 'Hydrophobe':
                maccs2.loc[i, 'Hydrophobe'] += 1
            elif feats[j].GetFamily() == 'LumpedHydrophobe':
                maccs2.loc[i, 'LumpedHydrophobe'] += 1
```
